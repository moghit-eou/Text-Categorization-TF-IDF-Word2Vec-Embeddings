```python
import warnings
import pandas as pd
import re
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("./data/high_dataset.csv")
data[ data["label"] == "job_applicant"].head()
```

```python
train_df , test_df = train_test_split (
        data , 
        test_size = 0.2 , 
        random_state = 42 , 
        stratify = data["label"] 
    )
```

```python
class_counts  = train_df["label"].value_counts()
```

```python
plt.figure(figsize = (10 , 3 ) )
sns.countplot( data = train_df ,  x = "label" , order = class_counts.index , palette = "Set2")
plt.show()
```

```python
import re

def tokenizer(text):
    text = text.lower()
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r',' , ' ' , text)
    text = re.sub(r'\.' , ' ' , text ) 
    text = re.sub(r'-' , ' ' , text ) 
    return text
```

```python
train_df["clean_text"] = train_df["text"].apply(tokenizer)
```

```python
label_mapping = {
    "customer" : 0 , 
    "co_worker" : 1 ,
    "job_applicant" : 2
}

train_df["label_encoding"] = train_df["label"].map(label_mapping)
```

```python
test_df["clean_text"] = test_df["text"].apply(tokenizer)
test_df["label_encoding"]  = test_df["label"].map(label_mapping)
```

```python
test_df
```

```python
X = train_df["clean_text"]
y = train_df["label"]
```

```python
X_train , X_test , y_train , y_test = train_test_split ( 
    X , 
    y , 
    test_size = 0.3 , 
    random_state = 42 ,
    stratify = y )
```

```python
tfidf_vectorizer = TfidfVectorizer(
    max_features = 30 , 
    ngram_range = ( 1 , 2 ),
    stop_words = "english"
)

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

X_val_tfidf = tfidf_vectorizer.transform(X_test)
X_test_tfidf = tfidf_vectorizer.transform(test_df["clean_text"])

```

```python
model = LogisticRegression(max_iter=20)
model.fit(X_train_tfidf, y_train)
```

```python
y_pred = model.predict(X_val_tfidf)
```

```python
cm = confusion_matrix (y_test , y_pred , labels = model.classes_ )
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()

```

```python
cm 
```

```python
accuracy = np.trace ( cm ) / np.sum ( cm )
accuracy
```

```python
def predict(text):
    text = tokenizer(text)
    text_idf = tfidf_vectorizer.transform([text])
    return model.predict(text_idf)
```
